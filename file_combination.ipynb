## Adding 4th column named as Review Date

```python
import re
from datetime import datetime
review_date = []
for review_dt in new_df['Review Date']:
  # check if the row contains a date in the format of dd-mm-yyyy
  match = re.search(r'(\d{2}-\d{2}-\d{4})', review_dt)
  if match:
    # Extract date using regex
    match = re.search(r'\d{2}-\d{2}-\d{4}', review_dt)
    date_str = match.group()

    # Convert date to datetime object
    date_obj = datetime.strptime(date_str, '%d-%m-%Y')

    # Format datetime object
    formatted_date = date_obj.strftime('%Y-%m-%d %H:%M:%S')
    review_date.append(formatted_date)
  else:
    # check if the row contains a date in the format of yyyy-mm-dd HH:MM
    match = re.search(r'(\d{4}-\d{2}-\d{2} \d{2}:\d{2})', review_dt)
    if match:
        date_str = match.group(1)
        # add :00 seconds to the date string to make it yyyy-mm-dd HH:MM:00
        date_str += ':00'
        review_date.append(date_str)
print(len(review_date))   
output_shoppe_belsimpel['Review_Date'] = review_date
output_shoppe_belsimpel
```

## now 5th column created 

```python
#Review Text
from bs4 import BeautifulSoup
review_Text = []

for review_text1, procon in zip(new_df['Review Text'], new_df['Pro_Con Review']):
  pattern = r'<.*?>' # This pattern matches any HTML tag
  if re.search(pattern, str(review_text1)):
    #print("In if statement")
    soup = BeautifulSoup(review_text1, 'html.parser')
    text_content = ''
    for div in soup.find_all('div'):
      text_content += div.get_text().strip() + "'.'"

      #print(text_content)
    review_Text.append(text_content)
  elif type(procon) == str :
    # split the string based on the "Pluspunt" keyword
    combined_review = ''
    plus_parts = procon.split("Pluspunt")[1:]
    # extract the words after "Pluspunt" to create a list of Pros
    pros = [p.split("Minpunt")[0] for p in plus_parts]
    pros = ', '.join(pros)
  
    min_parts = procon.split("Minpunt")[1:]
    # extract the words after "Minpunt" to create a list of Cons
    cons = [c.strip() for c in min_parts]
    cons = ', '.join(cons)
    if review_text1 or review_text1 != 0 :
      combined_review +=str(review_text1)
    if pros:
      combined_review+=" '.' Pros: "+pros
    if cons:
      combined_review += " '.' Cons: " + cons
    review_Text.append(combined_review)
  else:
    review_Text.append(review_text1)
output_shoppe_belsimpel['Review Text'] = review_Text
output_shoppe_belsimpel
```
